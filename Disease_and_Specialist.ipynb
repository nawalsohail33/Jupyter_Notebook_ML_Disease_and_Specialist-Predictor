```python
import random
import csv

# Medical data with specialists
medical_data = {
    "Cardiologist": {
        "Heart Disease": [
            "chest pain", "shortness of breath", "fatigue", "palpitations", "dizziness"
        ],
        "Hypertension": [
            "high blood pressure", "headache", "blurred vision", "chest discomfort"
        ]
    },
    "Neurologist": {
        "Migraine": [
            "headache", "nausea", "sensitivity to light", "throbbing pain", "blurred vision"
        ],
        "Epilepsy": [
            "seizures", "loss of consciousness", "confusion", "staring spells"
        ]
    },
    "Pulmonologist": {
        "Asthma": [
            "wheezing", "cough", "chest tightness", "shortness of breath"
        ],
        "Pneumonia": [
            "cough with phlegm", "fever", "chills", "difficulty breathing"
        ]
    },
    "Gastroenterologist": {
        "Food Poisoning": [
            "vomiting", "diarrhea", "abdominal pain", "nausea", "fever"
        ],
        "Irritable Bowel Syndrome": [
            "bloating", "gas", "diarrhea", "abdominal cramps"
        ]
    },
    "Psychiatrist": {
        "Depression": [
            "sadness", "loss of interest", "fatigue", "insomnia", "low self-esteem"
        ],
        "Anxiety Disorder": [
            "nervousness", "restlessness", "rapid heartbeat", "trouble concentrating"
        ],
        "Bipolar Disorder": [
            "mood swings", "high energy", "depression", "impulsivity"
        ]
    },
    "Psychologist": {
        "Stress Disorder": [
            "irritability", "tension", "trouble sleeping", "lack of focus"
        ],
        "Behavioral Disorder": [
            "impulsive behavior", "lack of discipline", "aggression", "difficulty in school"
        ]
    },
    "Dermatologist": {
        "Skin Allergy": [
            "itching", "red rash", "dry skin", "swelling", "flaky skin"
        ],
        "Acne": [
            "pimples", "oily skin", "blackheads", "inflammation", "skin bumps"
        ]
    },
    "Pediatrician": {
        "Common Cold (Children)": [
            "runny nose", "sneezing", "fever", "cough", "sore throat"
        ],
        "Ear Infection": [
            "ear pain", "trouble hearing", "crying in infants", "fever", "fluid from ear"
        ]
    },
    "Rheumatologist": {
        "Arthritis": [
            "joint pain", "swelling", "stiffness", "limited motion"
        ],
        "Lupus": [
            "fatigue", "joint pain", "rash", "fever", "swollen lymph nodes"
        ]
    }
}

# Helper function to generate symptom descriptions
def generate_symptom_description(symptoms):
    selected = random.sample(symptoms, random.randint(2, min(4, len(symptoms))))
    return " and ".join(selected)

# Generate synthetic dataset
rows = []
num_samples = 300  # Increase size for diversity

for _ in range(num_samples):
    specialist = random.choice(list(medical_data.keys()))
    condition = random.choice(list(medical_data[specialist].keys()))
    symptom_list = medical_data[specialist][condition]
    symptom_text = generate_symptom_description(symptom_list)
    rows.append([symptom_text, condition, specialist])

# Save to CSV
with open("C:/Users/HP/Documents/Programming For AI/project AI/project AI/project AI/synthetic_medical_dataset.csv", "w", newline='', encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["symptoms", "condition", "specialist"])
    writer.writerows(rows)

print(f"Dataset generated with {len(rows)} entries and saved as 'synthetic_medical_dataset.csv'")
```
